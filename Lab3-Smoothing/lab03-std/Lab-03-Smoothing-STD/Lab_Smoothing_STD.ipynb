```python
img = rgb2gray(io.imread('bird.jpg'))
noisy_img = random_noise(img, mode='s&p', amount=0.05)
```

```python
'''
(1) Your own median filter algorithm
'''

def myMedian(img):
    temp=np.zeros(img.shape)
    x=3
    y=3
    for i in range (0,img.shape[0]-2,1):
        for j in range (0,img.shape[1]-2,1):
#                 temp[i][j]=sorted(img[i:i+3,j:j+3].flatten())[4]
                temp[i+1][j+1]=np.median(img[i:i+3,j:j+3])
                
                
    return temp            
                      
    
```

```python
result=myMedian(noisy_img)
show_images([img,noisy_img,result],["original","noisy","My median"])
```

```python
'''
(2) skimage median filter
'''

result=median(noisy_img)
show_images([img,noisy_img,result],["original","noisy","skimage median"])
```

```python
#Comparison between skimage median & my median


show_images([noisy_img,median(noisy_img),myMedian(noisy_img)],["noisy","skimage median","My median"])

```

```python
'''
(3) Gaussion Filters
'''
from skimage.filters import gaussian

result1=gaussian(noisy_img,sigma=8)
result2=gaussian(noisy_img,sigma=3)
result3=gaussian(noisy_img,sigma=0.2)

show_images([img,noisy_img,result1,result2,result3],["original","noisy","gaussian filter with sigma 8","gaussian filter with sigma 3","gaussian filter with sigma 0.2"])

```
